```python
print('======= Menghitung nilai n jika x = (-1+np.sqrt(5))=======')
# Diketahui x memenuhi x**2+x-1=0.
# x**2=1-x
# Tentukan nilai n yang memenuhi x**2024-x**2025 = (x(1-x))**n
# Dengan menghitung secara manual di dapat nilai x yaitu
# x1 = (-1 + np.sqrt(5))/2
# x2 = (-1 - np,sqrt(5))/2
# 1. Import library
import numpy as np
# 2. Masukan nilai x
x_025 = (-1 + np.sqrt(5))/2
# 3. Hitung luas kiri (lhs) dari persamaan x**2024 - x**2025 = (x(1-x))**n dengan menggunakan logaritma
lhs_025 = 2024* np.log(np.abs(x_025)) + np.log(np.abs(1- x_025))
# 4. Hitung ruas kanan (rsh) dari persamaan x**2024 - x**2025 = (x(1-x))**n
rsh_025 = np.abs(x_025*(1-x_025))
# 5. Hitung nilai n
n_025 = lhs_025/np.log(np.abs(lhs_025))
# 6. Menampilkan hasil
print(f"untuk nilai n yang memenuhi jika x = (-1 + np.sqrt(5))adalah : {n_025}")
print('====== Thanks you ======')
```
