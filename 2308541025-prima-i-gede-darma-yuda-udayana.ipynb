```python
import math

# Mengatur nilai n sesuai NIM
n_25 = 25
a_25 = 1000 * n_25
b_25 = 1000 * (n_25 + 1)

#Mencari bilangan prima pada rentang nilai a_25 hingga b_25
angka_prima_25 = []

for c_25 in range (a_25, b_25 + 1):
    if c_25 > 1:
        prima_25 = True
        for i_25 in range(2,int(math.sqrt(c_25)) + 1):
            if c_25 % i_25 == 0:
                prima_25 =  False
                break        
    if prima_25:
        angka_prima_25.append(c_25)
    
#Menampilkan hasil
print("Bilangan prima antara",a_25," sampai dengan",b_25,"adalah")
print(angka_prima_25)
```
