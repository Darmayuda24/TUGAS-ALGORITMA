```python
def find_line_equation_025(x1, y1, x2, y2):
    # Menghitung kemiringan m
    m = (y2 - y1) / (x2 - x1)
    
    # Menghitung intercept b
    b = y1 - m * x1
    
    return m, b
# Input dua titik
x1 = float(input("Masukkan x1: "))
y1 = float(input("Masukkan y1: "))
x2 = float(input("Masukkan x2: "))
y2 = float(input("Masukkan y2: "))
# Mendapatkan m dan b
m, b = find_line_equation_025(x1, y1, x2, y2)
print(f"Persamaan garis: y = {m}x + {b}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
# Tentukan nilai a, b, dan c
a = 8
b = 9
c = 11
# Fungsi untuk menghitung nilai y berdasarkan x
def hitung_y(x):
    return (c - a * x) / b
# Buat range nilai x
x_values = np.linspace(-10, 10, 400)
y_values = hitung_y(x_values)
# Visualisasikan garis
plt.plot(x_values, y_values, label=f'{a}x + {b}y = {c}')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid(color = 'gray', linestyle = '--', linewidth = 0.5)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Grafik Garis ax + by = c')
plt.legend()
plt.show()
```
